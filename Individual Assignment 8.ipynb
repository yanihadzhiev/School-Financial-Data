```python
#import pandas as alias pd
import pandas as pd
#import matplotlib as alias plt
import matplotlib.pyplot as plt

#read the data
revenueData = pd.read_csv('SSF_2014-2016_00A01_with_ann.csv')
taxData = pd.read_csv('STC_2016_00A2_with_ann.csv')

#merge the two datasets
merged = pd.merge(revenueData, taxData, on='Geographic area name').fillna(0)

#use fancy indexing to only select the columns needed
data = merged[['Year_x', 'Year_y', 'Tax Type', 'Amount ($1,000)', 'Total ependiture ($1,000)', 'Debt outstanding at end of fiscal year ($1,000)', 'Total revenue ($1,000)']]

#use masking to filter rows where "Tax Type" is equal to "1121" (meaning "total tax")
#and the year_x (from mixed file) is 2016
maskedData = data[(data['Tax Type'] == 1121) & (data['Year_x'] == 2016)]

#create scatter plot to compare (1)tax income and expenditure of schools
plt.scatter(maskedData['Amount ($1,000)'], maskedData['Total revenue ($1,000)'], label='Total Revenue', color='blue', alpha = 0.2)
#(2)tax income and outstanding debt
plt.scatter(maskedData['Amount ($1,000)'], maskedData['Debt outstanding at end of fiscal year ($1,000)'], label='Outstanding Debt', color='orange', alpha = 0.2)
# (3)tax income and total expenditures
plt.scatter(maskedData['Amount ($1,000)'], maskedData['Total ependiture ($1,000)'], label='Expenditure', color='green', alpha = 0.2)


#add title
plt.title('Tax Income and School Finance 2016')
#add x label
plt.xlabel('Tax Income ($1,000)')
#add y label
plt.ylabel('Total Revenue, debt, and expenditures ($1,000)')

#add legend in the lower right
plt.legend(loc='lower right')

#display results
plt.show()

maskedData
```
